## Distribution and Counts

First of all for the classification problem. We should count the number of samples each class. We use numpy filter by condition `a[conditions]` for example, `a[a > 2]` which select only data in `a` greater than `2`. By this, we use cross variable. We select `X` which has `y == {some class}` by this code `X[y == {class ID}]`

```python
select_feature = 0
select_class = 1

data = X[y==select_class]

print(data.shape)
```

`y` which is class `1` have only 30 samples, so `X` is filtered to shape `(30, 3)`. Then show distribution first

```python
# make widget

plt.figure(figsize=(6,4), dpi=120)

# make chart
plt.hist(X[:, 1],  # all samples with feature 1
         color='green'
         )

# make properties

# plt.xlabel(f'feature {select_feature}')
# plt.ylabel(f'class {select_class}')

# plt.title('Histogram plot')

# plt.grid()

# plt.text(x=-2, y=5, s='A Text on the plot', fontsize=12,)

plt.show()
```

Try box plot to show distribution 

```python
plt.boxplot(X) # all features
plt.ylabel('value')
plt.title('Box Plot of Feature 2')
plt.show()
```

Filtering `y` using numpy condition may redundant to operation. There is a function `np.unique` to do this for us

```python
values, counts = np.unique(y, return_counts=True)

print(f'Unique values: {values}')
print(f'Count samples: {counts}')
```

Plot sample counts

```python
color_list = ['blue', 'orange', 'red']
label_list = ['Class 0', 'Class 1', 'Class 2']

plt.bar(values, counts, color=color_list, label=label_list, width=.5)

# plt.xlabel('class value')
# plt.ylabel('counts')

# plt.xticks(values)

# plt.title('Sample counts')

# plt.legend()

# for value, count in zip(values, counts):
#     plt.text(value - 0.05, count+0.5, f'{count}', fontsize=12)

plt.show()
```

See the pie chart is not bad for counting plot

```python
plt.pie(counts, 
        # colors=color_list, 
        # labels=label_list, 
        # autopct='%1.1f%%',
        # explode=[0, 0, 0.1],
        # shadow=True,
        # startangle=90,
        # wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
        )

plt.title('Pie Chart')

# plt.legend(bbox_to_anchor = [1, 0.7])

plt.show()
```

## Scatter plot and Feature maps

Scatter plot help a lot to roughly measure the feature distribution for pre-checking the data before doing analysis in order to select classification models

```python
f1, f2 = X[:, 0], X[:, 2]
```

Try plot only 2 featurs in a  space

```python
plt.scatter(f1, f2, c=y)

# plt.xlabel('Feature 1')
# plt.ylabel('Feature 2')

# plt.title('Scatter Plot of Feature 1 and Feature 2')

# plt.grid()

plt.show()
```

If we have 3 features, it is better to observe 3 feature spaces to evaluate distributions. This we employ `plt.subplots` which put `nrows` and `ncols` to plot multi plots by grid location. This example shows 1 row and 3 columns of sub plots. Then access each of sub plot using `ax` with `[row][col]` index (Noted that 1 row has no row dim, we can call `ax[idx]` directly)

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

ax[0].scatter(X[:, 0], X[:, 1], c=y)
# ax[0].set_title('F0 - F1')
# ax[0].set_xlabel('F0')
# ax[0].set_ylabel('F1')

ax[1].scatter(X[:, 1], X[:, 2], c=y)
# ax[1].set_title('F1 - F2')
# ax[1].set_xlabel('F1')
# ax[1].set_ylabel('F2')

ax[2].scatter(X[:, 0], X[:, 2], c=y)
# ax[2].set_title('F0 - F2')
# ax[2].set_xlabel('F0')
# ax[2].set_ylabel('F2')

fig.show()
```

Or use 3D plot, this may confuse people. It is good but not much recommend if it is jsut to show scatter plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')

# ax.set_xlabel('Feature 1')
# ax.set_ylabel('Feature 2')
# ax.set_zlabel('Feature 3')

# ax.set_title('A Title of Chart')

plt.show()
```

For advanced visualization, there is not only matplotlib can do. Some third-party libraries can customize matplotlibs tool to more easier and beautiful visualization like `seaborn`

```python
import seaborn as sns
import pandas as pd
```

If we are beginners of data science guys. we might use `pandas.DataFrame` to come up the dataset like

```python
data = pd.DataFrame({'F1': X[:, 0],
              'F2': X[:, 1],
              'F3': X[:, 2],
              'class': y})

data
```

`seaborn` assists one-function visualize different kind of customizable looks and feels like this pair plot that show mixed types of plot including histogram, scatter and subplots together

```python
sns.pairplot(data, hue='class', palette='bright')
```

Or use only some pair features to show scatters with histograms

```python
sns.jointplot(data=data, x="F1", y="F2", hue="class", palette="bright")
# sns.jointplot(data=data, x="F2", y="F3", hue="class", palette="bright")
# sns.jointplot(data=data, x="F1", y="F3", hue="class", palette="bright")
```

2D plotting is also allowed

```python
sns.kdeplot(
    data=data, x="F1", y="F3", hue="class", fill=True, palette="bright", alpha=.5
)
```
